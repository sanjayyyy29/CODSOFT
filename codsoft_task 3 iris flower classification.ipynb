```python
import pandas as pd
iris = pd.read_csv(r"C:/Users/krithika/Downloads/IRIS.csv")
```

```python
print(iris.describe())
```

```python
X = iris.drop("species", axis=1)
Y = iris["species"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, Y_train)
```

```python
import numpy as np
```

```python
new_data_point = np.array([[4, 0.2, 6, 2.2]])
```

```python
resulting_prediction = knn.predict(new_data_point)
```

```python
print("The prediction result for the given input is: {}".format(resulting_prediction))
```
